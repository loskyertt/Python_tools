```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# 1.数据处理

```python
df = pd.read_csv('../data/poi_counts.csv')
df
```

# 2.绘制分组柱状图

主题和字体配置。

```python
plt.style.use("../mpstyle/rose-pine-matplotlib/themes/rose-pine.mplstyle")   # 导入主题

plt.rcParams['font.sans-serif'] = ['SimHei']
# plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 绘制分组柱状图
labels = df["区域"]  # 区域名称
categories = ["医疗保健数量", "旅游景点数量", "交通设施数量", "生活服务数量"]
x = np.arange(len(labels))  # 生成一个与区域数量相同的序列，用作 x 轴坐标
width = 0.2  # 柱子的宽度
colors = ["#9BDFDF", "#FFE2CE", "#C4D8E9", "#BEBCDF"]  # 颜色列表

fig, ax = plt.subplots(figsize=(12, 6)) # 创建图形（像素：1200*600）和一个坐标轴

# 遍历类别，绘制柱状图，并添加数据标签
for i, category in enumerate(categories):
    bars = ax.bar(x=x + i * width, height=df[category],
                width=width, label=category, color=colors[i])

    # 在柱子上方添加数值标签
    for bar in bars:    # 遍历每个柱子
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, height,
                f'{height}', ha='center', va='bottom', fontsize=10)

# 设置图表信息
ax.set_xlabel("区域")
ax.set_ylabel("数量")
ax.set_title("各区域不同类别 POI 数量分布")
ax.set_xticks(x + width * 1.5)
ax.set_xticklabels(labels, rotation=45)
ax.legend()

plt.tight_layout()  # 自动调整子图参数，使其填充整个图表区域。
plt.show()  # 显示图表
```
