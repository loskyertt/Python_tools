```python
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1.处理 shp 数据

通过`geopandas`处理`shp`数据，和`pandas`处理数据很像，操作逻辑基本一样。

```python
gdf = gpd.read_file("../data/房价数据-shp/wuhan_house_price.shp", columns=['lng（经度', 'lat（纬度', 'prices（房'])
gdf.head()
```

```python
gdf = gdf.rename(
    columns={'lng（经度': 'lng', 'lat（纬度': 'lat', 'prices（房': 'prices'})
gdf.head()
```

# 2.绘制核密度估计图

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(10, 10))
plt.title("武汉市二手房价格分布")
sns.kdeplot(x=gdf['lng'], y=gdf['lat'], weights=gdf['prices'],
            fill=True, cmap='Blues', alpha=0.7)
plt.scatter(gdf['lng'], gdf['lat'], s=1, c='black', alpha=0.5)  # 绘制散点图
plt.xlabel('经度')
plt.ylabel('纬度')
plt.show()
```
